## Key Observations

- **Aggregate summary documents are highly effective.** Queries targeting category-, region-, or trend-level insights benefit directly from the purpose-built summary documents generated by `generate_documents.py`.
- **The model can hallucinate when context is incomplete.** For Q2, the absence of Copiers' sales figure in the retrieved docs led to an unverifiable 7.30% margin figure.
- **The model can misrank data from the retrieved context.** For Q8, all numbers were available but the model produced an incorrect ordering.
- **The model correctly respects data boundaries when retrieval is partial.** In Q7 (states) it appropriately admitted it could not rank all states given the limited context.
- **Trend and time-series questions need dedicated documents.** Static aggregate summaries are insufficient for trend comparison queries.

## Conclusion

The RAG pipeline achieved a **7/11 (64%) fully correct answer rate**, with 2 partial and 2 incorrect answers. Retrieval quality was strong for aggregate queries but inadequate for ranked or comprehensive cross-entity comparisons. Key improvements to target:

1. Add a **top-cities summary document** to `generate_documents.py` to support city-ranking queries.
2. Add **top-states summary document** for state comparison queries.
3. Add **category trend documents** (yearly breakdown per category) to support temporal category comparisons.
4. The confusion between *total profit* and *profit margin* in Q2 suggests prompt engineering could be improved to push the model to calculate ratios when both profit and sales are available in context.
